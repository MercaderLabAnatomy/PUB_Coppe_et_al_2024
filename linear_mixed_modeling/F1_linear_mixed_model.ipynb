```r
library(readxl)
library(dplyr)
library(tidyverse)
library(lmerTest) 
library(lme4)             
```

```r
data_animals_clean <- read_excel("/home/rstudio/raw/Injured_F1_no_outliers_IQR.xlsx")
head(data_animals_clean)
class(data_animals_clean)
```

```r
# Convert all columns except for the first 5 to numeric
data_animals_clean[] <- lapply(seq_along(data_animals_clean), function(i) {
  if (i > 5 && is.character(data_animals_clean[[i]])) {
    as.numeric(data_animals_clean[[i]])
  } else {
    data_animals_clean[[i]]
  }
})
```

```r
colnames(data_animals_clean)
```

```r
# lmer father and mother

# Directories for outputs
output_dir <- "/home/rstudio/TEST/lmer_uninjured_f1_father_mother_IQR_confint"

# Create directories if they don't exist
dir.create(output_dir, showWarnings = FALSE, recursive = TRUE)

# List of columns to analyze
columns_to_analyze <- c('Normalized_BW', 'Normalized_liver', 'Normalized_heart', 
                        'Normalized_LV_area_systole', 'Normalized_LV_area_diastole', 
                        'Normalized_LV_volume_systole', 'Normalized_LV_volume_diastole', 
                        'Normalized_IVS', 'Normalized_LVID_diastole', 'Normalized_LVID_systole', 
                        'Normalized_LVPW_diastole', 'Normalized_LV_Mass', 
                        'Normalized_LV_Mass_corrected', 'Normalized_LV_Vol_s', 
                        'Normalized_LV_Vol_d', 'Stroke_Volume', 'Ejection_Fraction', 
                        'Fractional_Shortening', 'Cardiac_Output', 'EF', 'FS', 
                        'E_velocity', 'A_velocity', 'E_vs_A') #, "EMH",'CMs_length','Camk1d','Triglycerides','Lipase','LDL','HDL','Total_Cholesterol','Free_Cholesterol','Calcium','Glucose','Free_fatty_acids')

# Initialize lists to store results
all_results <- list()

# Check for missing columns
missing_columns <- setdiff(columns_to_analyze, colnames(data_animals_clean))
if (length(missing_columns) > 0) {
  cat("Missing columns in the dataset:\n")
  print(missing_columns)
}

# Loop through each column and perform analyses
for (column in columns_to_analyze) {
  tryCatch({
    # Check if the column exists in the dataset
    if (!column %in% colnames(data_animals_clean)) {
      cat("Skipping column:", column, "as it is not found in the dataset.\n")
      next
    }
    
    # formula
    formula <- as.formula(paste(column, "~ Group + (1|Father) + (1|Mother)"))
    model <- lmer(formula, data = data_animals_clean)
    
    # Extract model summary and confidence intervals
    model_summary <- summary(model)
    model_confint <- confint(model)
    
    # Combine model summary coefficients and confidence intervals
    model_df <- as.data.frame(model_summary$coefficients)
    model_df$Variable <- column
    model_df$Term <- rownames(model_df)
    
    # Add confidence intervals to the model summary dataframe
    confint_df <- as.data.frame(model_confint)
    confint_df$Term <- rownames(confint_df)
    
    # Merge confidence intervals with model summary data
    combined_model_df <- merge(model_df, confint_df, by = "Term", all.x = TRUE)
    
    # Rename confidence interval columns for clarity
    colnames(combined_model_df)[(ncol(combined_model_df)-1):ncol(combined_model_df)] <- c("Lower_CL", "Upper_CL")
    
    # Store results
    all_results[[column]] <- combined_model_df
    
  }, error = function(e) {
    cat("An error occurred for column: ", column, "\n", "Error message: ", e$message, "\n")
  })
}

# Combine all the results into a single data frame
combined_results <- bind_rows(all_results)

# Reorder columns
combined_results <- combined_results %>%
  select(Variable, Term, everything())

# Write the combined results to CSV file
write.csv(combined_results, file.path(output_dir, "combined_lmer_summary.csv"), row.names = FALSE)

cat("LMER analysis and confidence intervals saved to:", file.path(output_dir, "combined_lmer_summary.csv"), "\n")
```
